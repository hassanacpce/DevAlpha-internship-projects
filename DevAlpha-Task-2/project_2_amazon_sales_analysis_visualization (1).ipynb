```python
#Top 10 Product Categories
top_categories = df["category"].value_counts().head(10)

plt.figure(figsize=(12,6))

top_categories.sort_values().plot(kind="barh", color="steelblue")

plt.title("Top 10 Product Categories", fontsize=14)
plt.xlabel("Number of Products")
plt.ylabel("Category")

plt.tight_layout()
plt.show()
```

```python
#Top 10 Product
top_categories = df["category"].value_counts().head(10)

short_names = [cat.split("|")[-1] for cat in top_categories.index]

plt.figure(figsize=(10,5))

plt.bar(short_names, top_categories.values)

plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
#Rating Distribution
plt.figure(figsize=(8,5))

sns.histplot(df["rating_numeric"].dropna(), bins=10)

plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.show()
```

```python
#Actual Price vs Discounted Price
plt.figure(figsize=(8,5))

plt.scatter(df["actual_price"], df["discounted_price"])

plt.title("Actual Price vs Discounted Price")
plt.xlabel("Actual Price")
plt.ylabel("Discounted Price")

plt.show()
```

```python
#Product Ratings
plt.figure(figsize=(6,4))

sns.boxplot(x=df["rating_numeric"].dropna())

plt.title("Product Ratings")

plt.show()
```

```python
#Correlation Heatmap

numeric_df = df.select_dtypes(include="number")

plt.figure(figsize=(6,5))

sns.heatmap(numeric_df.corr(), annot=True, cmap="Blues")

plt.title("Correlation Heatmap")

plt.show()
```

```python
#Top 10 Highest Rated Products

top = df.sort_values(by="rating_numeric", ascending=False).dropna(subset=["rating_numeric"]).head(10)

top["short_name"] = top["product_name"].apply(lambda x: x[:60] + "..." if len(str(x)) > 60 else x)

plt.figure(figsize=(12,6))

plt.barh(top["short_name"], top["rating_numeric"], color="steelblue")

plt.title("Top 10 Highest Rated Products")
plt.xlabel("Rating")
plt.ylabel("Product Name")

plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```
